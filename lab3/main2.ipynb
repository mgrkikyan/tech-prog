**Задание:**  

Записать определение производной функции одной вещественной переменной.
Доказать, что функция

$$

f(x) =
\begin{cases}
\sqrt{|x|} \cos \left(\frac{1}{x^2}\right), & \text{если } x \neq 0 \\
0, & \text{если } x = 0
\end{cases}

$$

непрерывна, но не дифференцируема в точке $x=0$

---




---

Производная функции одной вещественной переменной — это предел отношения приращения функции к приращению аргумента, когда последнее стремится к нулю. 

Математически это записывается следующим образом:

$$
f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}

$$

где:
- $f'(x)$ — производная функции $f$ в точке $x$,
- $\Delta x$ — приращение аргумента,
- $f(x + \Delta x) - f(x)$ — соответствующее приращение функции.

Производная характеризует скорость изменения функции в данной точке.

---



### Доказательство непрерывности функции

Функция $ f(x) $ определена кусочно:

$$
f(x) =
\begin{cases}
\sqrt{|x|} \cos \left(\frac{1}{x^2}\right), & \text{если } x \neq 0 \\
0, & \text{если } x = 0
\end{cases}

$$

Проверим непрерывность в точке $ x = 0 $. Для этого рассмотрим пределы функции слева и справа от этой точки:

$$
\lim_{x \to 0^-} f(x) = \lim_{x \to 0^+} f(x).

$$

Поскольку $ \sqrt{|x|} \rightarrow 0 $ при $ x \rightarrow 0 $, а $ \cos\left(\frac{1}{x^2}\right) $ ограничен на интервале $ [-1, 1] $, произведение этих функций стремится к нулю:

$$
\lim_{x \to 0} \sqrt{|x|} \cos\left(\frac{1}{x^2}\right) = 0.

$$

Таким образом,

$$
\lim_{x \to 0^-} f(x) = \lim_{x \to 0^+} f(x) = 0 = f(0),

$$

что означает, что функция непрерывна в точке $ x = 0 $.

### Доказательство отсутствия дифференцируемости

Теперь проверим существование производной в точке $ x = 0 $. Рассмотрим разностное отношение:

$$
\frac{f(h) - f(0)}{h} = \frac{\sqrt{|h|} \cos\left(\frac{1}{h^2}\right) - 0}{h} = \frac{\sqrt{|h|} \cos\left(\frac{1}{h^2}\right)}{h}.

$$

Для исследования предела этого выражения при $ h \to 0 $, упростим выражение:

$$
\frac{\sqrt{|h|} \cos\left(\frac{1}{h^2}\right)}{h} = \frac{\cos\left(\frac{1}{h^2}\right)}{\sqrt{h}}.

$$

При стремлении $ h $ к нулю, знаменатель $ \sqrt{h} $ стремится к нулю, а числитель $ \cos\left(\frac{1}{h^2}\right) $ колеблется между $ -1 $ и $ 1 $. Это означает, что предел данного выражения не существует, поскольку он зависит от значений аргумента $ h $, стремящегося к нулю.

Таким образом, функция $ f(x) $ не имеет производной в точке $ x = 0 $.
