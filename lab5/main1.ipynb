```python
import numpy as np
import matplotlib.pyplot as plt

# A = nр.array([[a11, a12],[a21, а22]])
# Заданная матрица A
A = np.array([[3, 1],
              [1, 3]])

# Нахождение собственных значений и векторов
eigenvalues, eigenvectors = np.linalg.eig(A)

# Получили собственные векторы и собственные значения
b1 = eigenvectors[:, 0]
b2 = eigenvectors[:, 1]

# График с двумя панелями
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Уравнение единичной окружности
t = np.linspace(0, 2*np.pi, 100)
x = np.cos(t)
y = np.sin(t)

# Применение матрицы A к координатам точки
u = A[0, 0]*x + A[0, 1]*y
v = A[1, 0]*x + A[1, 1]*y

# Отображение эллипса (преобразованной окружности)
axes[1].plot(u, v, 'k-', label="Образ окружности")

# Вектор Ab1 и Ab2
Ab1 = A @ b1
Ab2 = A @ b2

# Добавление векторов на второй график
axes[1].quiver(0, 0, Ab1[0], Ab1[1], color='red', scale_units='xy', scale=1, width=0.007, label=r'$Ab_1$')
axes[1].quiver(0, 0, Ab2[0], Ab2[1], color='blue', scale_units='xy', scale=1, width=0.007, label=r'$Ab_2$')

# Настройка второго графика
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$y$')
axes[1].axis('equal')
axes[1].legend(loc='best')

# Рисунок первой панели: Исходная окружность и собственные векторы
axes[0].plot(x, y, 'k-', label="Единичная окружность")

# Собственные векторы
axes[0].quiver(0, 0, b1[0], b1[1], color='red', scale_units='xy', scale=1, width=0.007, label=r'$b_1$')
axes[0].quiver(0, 0, b2[0], b2[1], color='blue', scale_units='xy', scale=1, width=0.007, label=r'$b_2$')

# Настройки первого графика
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
axes[0].axis('equal')
axes[0].legend(loc='best')

# Показываем график
plt.tight_layout()  # оптимизация размещения панелей
plt.show()

eigenvalues, eigenvectors

```
