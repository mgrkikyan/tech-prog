```python
import numpy as np
import matplotlib.pyplot as plt
N = 10000
n = 1800
a = 0
b = 1
k = 40
A = (a + b) / 2
D = ((b - a) ** 2) / 12

random_numbers = np.random.uniform(a, b, size=(N, n))

Sn = random_numbers.sum(axis=1)
An = n * A
Dn = n * D
X = (Sn - An) / np.sqrt(Dn)


x = np.linspace(-4, 4, 1000)
y = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*x**2)


plt.plot(x, y, 'r-', lw=3)
plt.hist(X, density=True, bins=k)
```
